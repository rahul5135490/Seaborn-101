```python
import seaborn as sns
df=sns.load_dataset("titanic")
```

```python
df.head(5)
```

# Countplot

```python
sns.countplot(x="sex",data=df)
# df=dataframe
# x=column
```

```python
sns.countplot(x="adult_male",data=df)
# df=dataframe
# x=column
```

```python
sns.countplot(x="class",data=df)
# df=dataframe
# x=column
```

```python
sns.countplot(x="age",data=df)
# df=dataframe
# x=column
```

```python
sns.countplot(x="sex",hue="survived",data=df)
# df=dataframe
# x=column
```

```python
sns.countplot(x="adult_male",hue="survived",data=df)
# df=dataframe
# x=column
```

```python
sns.countplot(x="age",hue="survived",data=df)
# df=dataframe
# x=column
```

# KDE plot
# Continuous data

```python
sns.kdeplot(x="age",data=df,color="black")
# df=dataframe
# x=column
```

# Distribution plot(continuous data)

```python
sns.distplot(x="age",data=df)
# df=dataframe
# x=column
```

```python
# Scatter plot
df=sns.load_dataset("iris")
```

```python
df
```

```python
sns.scatterplot(x="sepal_length",y="petal_length",data=df,hue="species")
```

```python
# joint plot
sns.jointplot(x="sepal_length",y="petal_length",data=df,hue="species")
```

```python
corr=df.corr()
sns.heatmap(corr)
```
